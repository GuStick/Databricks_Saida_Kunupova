## **Streaming Query**

```python
# Using DDL schema
my_schema = """
                order_id INT,
                customer_id INT,
                order_date DATE,
                amount DOUBLE
     """       
```

```python
df_batch = spark.read.format("csv")\
            .option("header", "true")\
            .schema(my_schema)\
            .load("/Volumes/pyspark_cata/source/db_volume/StreamSource/")
    
display(df_batch)
```

```python
df = spark.readStream.format("csv")\
            .option("header", "true")\
            .schema(my_schema)\
            .load("/Volumes/pyspark_cata/source/db_volume/StreamSource/")
    
display(
    df,
    checkpointLocation="/Volumes/pyspark_cata/source/db_volume/StreamSource/checkpoint_display/"
)
```

### **Streaming Query**

```python
df.writeStream.format("delta")\
    .option("checkpointLocation", "/Volumes/pyspark_cata/source/db_volume/streamSink/checkpoint")\
    .option("mergeSchema", "True")\
    .trigger(once = True)\
     #.trigger(processingTime= "10 seconds")\
    .start("/Volumes/pyspark_cata/source/db_volume/streamSink/data")

```
